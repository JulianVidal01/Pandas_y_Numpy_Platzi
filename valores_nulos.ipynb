```python
import pandas as pd
import numpy as np
```

```python
dict = {'Col1':[1,2,3,np.nan],'Col2':[4,np.nan,6,7],'Col3':['a','b','c',None]}
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

```python
df.isnull() #Manejo de valores nulos
```

```python
df.isnull()*1 #Manejar nulos como numericos 1 donde hay valores nulos
```

```python
df.fillna('Missing') #Cade que exista un valor nulo como Missing
```

```python
df.fillna(df.mean()) #Sacar el valor medio
```

```python
df.interpolate() #Crear un valor como si fuera una serie
```

```python
df.dropna() #Borrar los valores nulos
```
