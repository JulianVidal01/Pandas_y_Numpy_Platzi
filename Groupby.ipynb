```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/drive/MyDrive/PLATZI/Curso Básico de Manipulación y Transformación de Datos con Pandas y NumPy/bestsellers-with-categories.csv',sep=',')
df_books.head(2)
#Para serpara un archivo distinto a comas se utiliza el siguiente codigo sep='' por el digito de separación
#Header sirve para informar en que posición se encuentra el archivo, en caso de que no tenga header se coloca none.
```

```python
df_books.groupby('Author').count()
#Agrupar por Author y mostrar el conteo de los datos de las demás columnas
```

```python
df_books.groupby('Author').median()
#Agrupar por Author y mostrar la media de los datos de las demás columnas
```

```python
df_books.groupby('Author').sum().loc['William Davis']
#Podemos usar loc y acceder a un dato específico del DataFrame. Agrupar por autor y mostrar la suma de los valores de las demás columnas para William Davis
```

```python
df_books.groupby('Author').sum().reset_index()
#Agrupar por author y mostrar la suma de los valores de las demás columnas. Colocar los índices que el DataFrame trae por defecto
```

```python
df_books.groupby('Author').agg(['min','max']) 
#La función agg() permite aplicar varias funciones al DataFrame una vez agrupado
# según una columna específica. Agrupar por Author y mostrar el mínimo y máximo de las demás columnas
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'}) #Agrupar por Author, obtener el mínimo y máximo de la columna ‘Reviews’ y sumar los valores de la columna ‘User Rating’
```

```python
df_books.groupby(['Author','Year']).count() #Agrupar por ‘Author - Year’ y contar los valores de las demás columnas
```
